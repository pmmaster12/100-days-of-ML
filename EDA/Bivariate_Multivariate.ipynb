# 1 Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(data = tips, x= 'total_bill',y='tip', hue = 'smoker', style = 'sex', size = 'size')
# hue is like add color code to data points based on a specific numerical or categorical column
```

# 2. Bar Plot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(data=titanic, x= 'Pclass', y= 'Age', hue = 'Sex')
# better practise expecting categorical feature in x axis)
```

# 3. Boxplot (categorical - numerical)

```python
sns.boxplot(data=titanic, x= 'Pclass', y= 'Age', hue = 'Sex')
```

# 4. Distplot ( Numerical - Categorical)

```python

sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

```python
sns.displot(data=tips, x='total_bill', hue = 'sex', kde = True)
```

# 5. HEATMAP ( CAT - CAT)

```python
numeric_tips = tips.select_dtypes(include=['number'])

# 2. Calculate the correlation matrix
corr_matrix = numeric_tips.corr()
```

```python
corr_matrix
```

```python
sns.heatmap(data=corr_matrix, annot=True, cmap='coolwarm')
```

# 6. Clustermap (CAT - CAT)

```python
tips.info()
```

```python
day_time_counts = pd.crosstab(index=tips['day'], columns=tips['time'])
```

```python
day_time_counts
```

```python
sns.clustermap(data=day_time_counts, annot=True, fmt='d', cmap='Blues')
```
