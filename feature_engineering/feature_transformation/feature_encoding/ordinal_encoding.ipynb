## Use for encoding or handling categorical column in which classes have relationship or order in between them

```python
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.iloc[:,0:2], df['purchased'], test_size =0.3, random_state=42)
```

```python
X_train.shape , X_test.shape
```

```python
X_train['review'].value_counts(), X_train['education'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
```

```python
X_train
```

```python
oe.categories_
```

```python

from sklearn.preprocessing import LabelEncoder
```

```python

le = LabelEncoder()
```

```python
le.fit(Y_train)
```

```python
le.classes_
```

```python
Y_train = le.transform(Y_train)
Y_test = le.transform(Y_test)
```
