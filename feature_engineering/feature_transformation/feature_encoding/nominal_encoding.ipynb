# 3. One-Hot Encoding using Scikit-Learn

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python

from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False, dtype=int)
```

```python
ohe.fit(X_train[['fuel','owner']])
```

```python
X_train_new = ohe.transform(X_train[['fuel', 'owner']])
```

```python
X_train_new
```

```python
X_test_new = ohe.transform(X_test[['fuel', 'owner']])
```

```python
X_train_new = pd.DataFrame(X_train_new, columns=ohe.get_feature_names_out(['fuel', 'owner']), index=X_train.index)
print(X_train_new)
```

```python
X_test_new = pd.DataFrame(X_test_new, columns=ohe.get_feature_names_out(['fuel', 'owner']), index= X_test.index)
print(X_test_new)
```
